```r
library(tidyverse)
```

```r
stuff <- read_csv("data/players.csv") 
```

```r
genders_data<- data.frame(genders <- c("Male", "Female", "Non-Binary", "Other", "Prefer not to say", "Agender", "Two- Spirited"))

gender_dummies <- model.matrix(~ genders - 1, data = genders_data)
player_data <- cbind(gender_dummies, stuff)
player_data
```

```r
hours_gen_sp <- ggplot(stuff, 
                    aes(x = gender, y = played_hours)) +
  geom_point() +
  labs(x = "Age of Players", y = "Total Playing Time (hrs)", 
       title = "Gender and Playing Time", fill = "Gender") 
hours_gen_sp
```
